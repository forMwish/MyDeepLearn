```
from keras.datasets import reuters

(train_data, train_label), (test_data, test_label) = reuters.load_data(num_words=10000)
print(train_data.shape, test_data.shape)
print(test_label[0])
print(type(test_data), type(test_data[0]))
```

```
def decode(sequence):
  word_index = reuters.get_word_index()
  print(type(word_index.items()))
  
  index_word = dict({(index, word) for (word, index) in word_index.items()})
  print(index_word)
  return " ".join(index_word.get(i - 3, "?") for i in sequence)

print(decode(train_data[0]))
```

```
import numpy as np

def one_hot(seqences, dimention=10000):
  result = np.zeros((len(seqences), dimention), dtype=np.float32)
  for i, data in enumerate(seqences):
    result[i, data] = 1.
  return result

train_x = one_hot(train_data)
test_x = one_hot(test_data)
train_y = one_hot(train_label, dimention=46)
test_y = one_hot(test_label, dimention=46)

train_x_val = train_x[:1000]
train_x_partial = train_x[1000:]
train_y_val = train_y[:1000]
train_y_partial = train_y[1000:]
print(train_y_val.shape)
print(train_x_val.shape)
```

```
from keras import models
from keras import layers
from keras import optimizers
from keras import losses
from keras import metrics

model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(64, activation="relu"))
model.add(layers.Dense(46, activation='softmax'))

model.compile(optimizer = optimizers.RMSprop(),
    loss = losses.categorical_crossentropy,
    metrics = metrics.categorical_accuracy)
```

```
history = model.fit(train_x_partial,
          train_y_partial,
          epochs=20,
          batch_size=512,
          validation_data=(train_x_val, train_y_val))
```
